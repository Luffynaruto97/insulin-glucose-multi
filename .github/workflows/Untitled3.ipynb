```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
def model(y,t):
  dydt = -y + 1
  return dydt

y0 = 0

t = np.linspace(0,5)

y = odeint(model,y0,t)

plt.plot(t,y,'r--')
plt.xlabel('time')
plt.ylabel('y(t)')
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt 

def model(y,t):
    if t< 10:
     u=0
    else:
     u=2
    dydt = 0.2*(-y + u)
    return dydt
y0 = 1

t= np.linspace(0,40)

y = odeint(model,y0,t)

plt.plot(t,y, 'g:')
plt.xlabel ('time')
plt.ylabel ('y(t)')
plt.show()

```
